```python
# selenium은 사용자가 볼 수 있는 만큼만 볼 수 있는 라이브러리
from selenium import webdriver as web
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver import ActionChains
import time
```


- 유투브 커스텀드 태그 : yt~

```python
# 브라우저를 실행 > 유튜브채널 사이트로~
driver = web.Chrome()
driver.get('https://www.youtube.com/watch?v=y9VpsNEdHQ8')
time.sleep(5)
```

```python
# 댓글 정보를 수집(스크롤을 안내려서 댓글이 수집 안됨)
# 클라이언트와 서버가 정보를 주고 받는 방식
# 동기 통신 : 
# 비동기 방식 : 한번에 HTML 화면이 보이려면 시간이 오래걸리기 때문에 일단 먼저 보이는 부분만 불러와서 화면에 띄우는 방식
review_li = driver.find_elements(By.ID, 'content-text')


for review in review_li :
    print(review.text)
```

```python
# 수집할 데이터의 길이 체크
len(review_li)
```

### 스크롤 내리는 방법
- pg dn : 페이지의 절반 정도만 내려감
- end : 페이지 끝까지 내려감
#### >> 키보드 END키를 활용!
- 전체 페이지를 움직이기 때문에 body태그를 접근
- body 태그에게 키보드의 END값을 전달 > body.send_keys(Keys.END)
#### [주의점]
- 반복 작업시, while문은 사용금지 > body는 계속 존재 > 무한 루프
- for문을 활용해서 많은 양의 반복을 통해 진행 추천!

```python
# 스크롤 끝까지 내리기
for i in range(50):
    body = driver.find_element(By.TAG_NAME, 'body')
    body.send_keys(Keys.END)
    time.sleep(0.5)
```
