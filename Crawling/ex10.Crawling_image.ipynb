```python
from selenium import webdriver as web
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time

# 파일 시스템을 위한 라이브러리(파일, 폴더를 생성, 삭제)
import os

# 이미지의 url값을 파일로 변형시켜주는 라이브러리
from urllib.request import urlretrieve
```

```python
# 1. 크롬으로 네이버창 오픈
driver = web.Chrome()
driver.get('https://www.naver.com/')
```

```python
# 2. 검색창에 원하는 검색 입력
# 2.1  검색까지 진행
search = driver.find_element(By.ID, 'query')
search.send_keys('손흥민')
search.send_keys(Keys.ENTER)
```

```python
# 3. 이미지 탭 클릭
img_tap = driver.find_element(By.CSS_SELECTOR, '#lnb > div.lnb_group > div > ul > li:nth-child(2) > a')
img_tap.click()
```

## html 이미지 src 경로 방식
1. 내부 방식(로컬)
2. 외부 방식(CDN)

```python
# 4. 이미지 태그들을 수집
# 이유는? 이미지 태그 속에 존재하는 src만 추출하기 위해서!
img_li = driver.find_elements(By.CSS_SELECTOR, '._image._listImage')
```

```python
# 나 img리스트의 0번째 데이터에서 src속성을 가져다줘!
# src만 담아줄 리스트 새로 제작
src_li = []
for img in img_li :
    
    # 속성을 수집할 때는 get_attribute('속성')
    src_li.append(img.get_attribute('src'))
```

```python
# 이미지 저장!
# 바탕화면에 이미지 폴더를 생성
# 바탕화면에 이미지라는 폴더가 없다면, 바탕화면에 이미지 폴더 만들어줘~
if not os.path.isdir('C:/Users/AI/GJAIschool/Web/Crawling/image') :
    # mk : make의 약자  / mkdir : directory를 만들어라.
    # rm : remove의 약자
    # 리눅스 명령어임
    os.mkdir('C:/Users/AI/GJAIschool/Web/Crawling/image')
```


- ../ : 폴더 한번 나가는 것
- / : 루트까지 나가는 것

```python
# url값을 이미지 폴더에 저장
cnt = 0
for i in src_li :
    urlretrieve(i, f'C:/Users/AI/GJAIschool/Web/Crawling/image/image{cnt}.jpg')
    cnt += 1
    time.sleep(2)
```

# 이미지가 중간에 깨지는 이유는?
- 이미지는 텍스트 파일보다 크기가 더 크기 때문에 
- 화면상에 스크롤을 통해서 더 많이 로딩을 받기!
- 1. 크롬 실행
- 2. 스크롤을 충분히 진행   // 화면 구성
- 3. img태그를 수집         // 태그 수집
- 4. img태그 속 src만 추출  // 데이터 가공
- 5. 파일로 저장            // 데이터 활용

```python
# 스크롤바 내리기
for i in range(100) :
    body = driver.find_element(By.TAG_NAME, 'body')
    body.send_keys(Keys.END)
    time.sleep(0.5)
```
