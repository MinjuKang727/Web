# 예외 처리
- 오류가 나기 전까지 실행하다 오류가 났을 때는 다른 코드를 실행!
- try = 오류가 나기 전까지 코드
- except = 오류가 났을 때 코드

```python
# 더보기를 반복해서 끝까지 수행
# 반복 횟수가 다 다르기 때문에, while문을 활용!
try :
    # 오류 나기 전,
    while True :
        btn = driver.find_element(By.CSS_SELECTOR, '#btn_more > span > a')
        btn.click()
        time.sleep(1)
except :
    # 오류 난 후,
    print('더보기 완료!')
```

```python
# 0. 라이브러리 추가
from selenium import webdriver as web
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
import pandas as pd

# 1. 크롬창 실행 후 > 한솥메뉴 사이트 요청
driver = web.Chrome()
driver.get('https://www.hsd.co.kr/menu/menu_list')

# 2. 더보기를 끝까지 누르는 로직
try :
    while True :
        see_more = driver.find_element(By.CSS_SELECTOR,'a.c_05')
        see_more.click()
        time.sleep(1)
except :
    print('더보기 클릭 완료!')
    
# 3. 가격, 메뉴명을 수집
menu_li = driver.find_elements(By.CSS_SELECTOR,'.item-text > h4.fz_03')
price_li = driver.find_elements(By.CSS_SELECTOR,'.item-price > strong')

# 4. 순수한 텍스트 정보만 추출 >> 비어있는 리스트에 텍스트만 담기
menu_ = []
price_ = []

for i in range(len(menu_li)) :
    menu_.append(menu_li[i].text)
    price_.append(price_li[i].text)
    
# 5. 데이터프레임 제작 >> 판다스
hansot = pd.DataFrame({
    '메뉴' : menu_,
    '가격' : price_
})
# 6. 한글이 깨지지 않고 csv 파일을 제작
hansot.to_csv('data/hansot_menu_list.csv', encoding = 'euc-kr')

# 7. 크롬창을 종료!
driver.close()
```

```python
## 신인수's 코딩 ##

from selenium import webdriver as wb
from selenium.webdriver.common.by import By
import time
import pandas as pd

#1. 크롬창 실행 후 > 한솥메뉴사이트 요청
driver = wb.Chrome()
driver.get('https://www.hsd.co.kr/menu/menu_list')

#2. 더보기를 끝까지 누르는 로직 작성
try :
    while True:
        btn = driver.find_element(By.CSS_SELECTOR,'#btn_more > span > a')
        btn.click()
        time.sleep(1)
except :
    print('더보기 완료')
    
#3. 가격, 메뉴명을 수집
title = driver.find_elements(By.CLASS_NAME ,'h.fz_03')
price = driver.find_elements(By.CLASS_NAME , 'item-price')

#4. 순수한 텍스트 정보만 추출 -> 비어있는 리스트에 텍스트만 담기
list0 = [(title[i],price[i]) for i in range(len(title))]
list_df = []
for i,j in list0:
    list_df.append((i.text, j.text))
    
#5. 데이터 프레임 제작 -> 판다스
data = pd.DataFrame(list_df, columns = ['메뉴', '가격'])
print(data)

#6. 한글이 깨지지 않고 csv파일을 제작
data.to_csv('한솥수정본.csv', encoding='euc-kr')

# 7. 크롬창을 종료!
driver.close()

# 리스트에 (메뉴,가격) 튜플을 담아서 DataFrame에 튜플이 담긴 리스트 데이터를 넣고, 칼럼명을 주었다.
```

# ActionChains
- ActionChains 사용법
1. 내가 액션을 줄 모든 요소들을 수집
2. ActionChains('브라우저') 호출
3. ActionChains('브라우저').기능(요소명).기능(요소명).perform()


```python
# 이벤트를 묶어서 실행하는 라이브러리
from selenium.webdriver import ActionChains

# 모바일 화면과 PC용 화면은 반응형 디자인으로 서로 다르게 구성되므로 반드시 PC 화면으로 브라우저를 열어서 동작을 수행하자!
```

```python
driver = web.Chrome()
driver.get('https://www.hsd.co.kr/')
```


- CSS_Selector 값 찾는 법

```python
# 0. 메뉴 탭 찾기!
menu = driver.find_element(By.CSS_SELECTOR, '#gnb > div.gnb_menu > ul > li:nth-child(3) > p > a')
# 1. 메뉴 탭에 마우스를 올리기!
# 2. 올린 상태에서 전체 메뉴를 클릭
allList = driver.find_element(By.CSS_SELECTOR, '#gnb > div.gnb_menu > ul > li:nth-child(3) > div > ul > li:nth-child(1) > a')
ActionChains(driver).move_to_element(menu).click(allList).perform()
```
