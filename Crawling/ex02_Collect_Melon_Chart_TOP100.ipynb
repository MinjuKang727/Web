```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
# 1. req를 활용해서 멜론차트 페이지 정보 요청해서 받아오기!
# 2. <Response [406]> = 통신에 실패(요청에서 처리-클라이언트) >> 브라우저로 안열어서
# 3. 에러코드가 400번 대(클라이언트), 500번대(서버) : 500번 대 에러는 못 가지고 오는 이야기
req.get('https://www.melon.com/chart/index.htm')
```


- 블록 입힌 부분 브라우저만 접속 허용됨

```python
# 브라우저인 척 속이는 작업이 필요
h = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36'}
```

```python
res = req.get('https://www.melon.com/chart/index.htm', headers = h)
soup = bs(res.text, 'lxml')
```

# 1. 노래 제목 수집

```python
# 노래 제목에 해당하는 태그 정보를 알아온 다음에,
# soup.select() 노래 제목만 수집

# title = soup.select("div.ellipsis rank01 > span > a") 라고 쓰면 클래스명에 공백이 있어서 검색 안됨
# (클래스에 공백이 있으면 클래스를 2개 가지고 있는 것!)
# title = soup.select("div.ellipsis.rank01 > span > a") 공백을 점(.)으로 대체해야 함
title = soup.select("div.ellipsis.rank01 > span > a")
for i in title :
    print(i.text)
    
    
#나는 아래 코드로 씀
# data = soup.select(".wrap_song_info div:first-child>span>a")
# for i in data :
#     print(i.text)
```
