```python
# webdriver = 브라우저 
from selenium import webdriver as wb

# Keys = 키보드
from selenium.webdriver.common.keys import Keys

# 요소로 검색
from selenium.webdriver.common.by import By
```




## 내 컴퓨터의 크롬과 같은 사양의 크롬 다운





```python
# 크롬을 실행하기 위해서는 크롬드라이버를 설치
# 사용자가 사용하는 크롬과 같은 같은 버전을 설치
# 브라우저가 업데이트 되면 크롬드라이버도 업데이트
# 작업 중인 파일과 크롬 드라이버를 같은 경로에 배치
```

```python
# 크롬창을 시작
# 웹 드라이버야 크롬 좀 실행 시켜줘~!
#  wb.Chrome("크롬 드라이버 경로 / chromedriver.exe")  ****반드시 대문자 C로 시작!!******
# 같은 경로 안에 있다면 경로를 생략이 가능
driver = wb.Chrome()
driver.get('https://www.google.com/')
```

```python
#.get('url') = 페이지 정보를 요청
driver.get('https://www.naver.com/')
# 여러개의 데이터를 검색할 때에는 class 선택자를 쓰고
# 한개의 데이터를 검색할 때에는 id선택자를 쓰는 것이 좋다.
# 우클릭 검색 단축키 : Ctrl + Shift + C
```

```python
# 검색창을 찾아오라고 명령!
# 아이디가 query인 요소 좀 찾아와 줘!
search = driver.find_element(By.ID, 'query')
```

```python
# 검색창에 키 값좀 보내줘~
# send_keys("전달값") = 값을 전달하는 명령
search.send_keys("광주날씨")
```

```python
# Enter키 입력
search.send_keys(Keys.ENTER)
```

```python
# 페이지 뒤로 가기
# HTML 영역이 아닌 브라우저 영역이기 때문에 driver에게 명령
driver.back()
```

```python
# search.send_keys("서울날씨")
# StaleElementReferenceException  발생!!
# 화면이 바뀌면 요소를 매번 새로 탐색
search = driver.find_element(By.ID, 'query')
search.send_keys("광주날씨")
```

```python
# 입력 버튼 찾기
btn = driver.find_element(By.ID, 'search_btn')
```

```python
# 버튼 클릭하기
# .click() : 요소를 클릭하는 메소드
btn.click()
```

```python
# 창을 종료
driver.close()
```
