```python
from selenium import webdriver as web
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
```

```python
# 1. 크롬 브라우저를 실행
driver = web.Chrome()

# 1.1 한솥 메뉴 사이트로 이동
driver.get('https://www.hsd.co.kr/menu/menu_list')
```

```python
# 상품명, 상품 가격 정보를 수집(글자만 출력)
# menu_li = driver.find_elements(By.TAG_NAME, 'h4')
menu_li = driver.find_elements(By.CLASS_NAME, 'h.fz_03') 
# 클래스가 여러개 주어져 있을 경우, 뒤에 오는 클래스 앞에는 .(dot)을 붙여 주어야 한다.

# price_li = driver.find_elements(By.CLASS_NAME, 'item-price')
# By.CSS_SELECTOR = 선택자를 직접 명시할 때 사용
# 주의점! 아이디: #(sharp)  / 클래스 : .(dot)
price_li = driver.find_elements(By.CSS_SELECTOR, '.item-price > strong')

menu_ = []
price_ = []

for i in menu_li :
    menu_.append(i.text)
for i in price_li :
    price_.append(i.text)

for i in range(len(menu_)) :
    print(menu_[i],price_[i])
```

```python
# 더 보기 버튼을 끝까지 누르는 로직

more = driver.find_element(By.CSS_SELECTOR, '#btn_more > span > a')

more.click()
```

# 예외 처리
- 오류가 나기 전까지 실행하다 오류가 났을 때는 다른 코드를 실행!
- try = 오류가 나기 전까지 코드
- except = 오류가 났을 때 코드

```python
# 더보기를 반복해서 끝까지 수행
# 반복 횟수가 다 다르기 때문에, while문을 활용!
try :
    # 오류 나기 전,
    while True :
        btn = driver.find_element(By.CSS_SELECTOR, '#btn_more > span > a')
        btn.click()
        time.sleep(1)
except :
    # 오류 난 후,
    print('더보기 완료!')
```

## 한 Cell에 완벽한 크롤링 코드 짜기

```python
# 0. 라이브러리 추가
from selenium import webdriver as web
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
import pandas as pd

# 1. 크롬창 실행 후 > 한솥메뉴 사이트 요청
driver = web.Chrome()
driver.get('https://www.hsd.co.kr/menu/menu_list')

# 2. 더보기를 끝까지 누르는 로직
try :
    while True :
        see_more = driver.find_element(By.CSS_SELECTOR,'a.c_05')
        see_more.click()
        time.sleep(1)
except :
    print('더보기 클릭 완료!')
    
# 3. 가격, 메뉴명을 수집
menu_li = driver.find_elements(By.CSS_SELECTOR,'.item-text > h4.fz_03')
price_li = driver.find_elements(By.CSS_SELECTOR,'.item-price > strong')

# 4. 순수한 텍스트 정보만 추출 >> 비어있는 리스트에 텍스트만 담기
menu_ = []
price_ = []

for i in range(len(menu_li)) :
    menu_.append(menu_li[i].text)
    price_.append(price_li[i].text)
    
# 5. 데이터프레임 제작 >> 판다스
hansot = pd.DataFrame({
    '메뉴' : menu_,
    '가격' : price_
})
# 6. 한글이 깨지지 않고 csv 파일을 제작
hansot.to_csv('data/hansot_menu_list.csv', encoding = 'euc-kr')

print('데이터 수집 완료')
```
