```python
# import 작업
# requests = 브라우저의 역할을 대신(요청, 응답 정보를 처리)
import requests as req
```

```python
# 1. 네이버 url을 통해서 사이트를 요청
# req야 네이버 페이즈 좀 가져와 줘! : req.get(URL)
# <Response [200]> : 통신에 성공했습니다.(안 넘어올 경우도 있음.)
# 200~300번 대 : OK!. 성공!
# 404(400번 대) : 사용자(클아이언트) 문제. 요청에 문제가 있는 것
# 500번 대 : 서버 문제
# ********request 결과가 200번대인지 꼭 확인하고 넘어가야한다!!!!!!!!!**********
res = req.get('https://naver.com')   # Response 안에 text, 응답, 통신, 에러 코드 등의 정보가 있음
```

```python
# 2. 응답받은 페이지에서 원하는 정보를 추출(카페, 블로그, ...)
# res : 응답 객체 (딕셔너리(res)의 키(text, 응답, 통신 등)와 값 같은 느낌)
print(res.text)   # html 소스처럼 보이는 문자열(out[] 확인해보면 작은따옴표('')로 시작)  >> 컴퓨터가 이해할 수 X(사용자를 위한 데이터)
```

# 컴퓨터가 알아먹을 수 있는 언어로 변형
## beautifulSoup 라이브러리
- 응답받은 텍스트 데이터를 html 데이터로 변형

```python
# 폴더 안에 들어 있는 라이브러리를 불러올 때 : from 패키지명 import 라이브러리명
# BeautifulStoneSoup(스톤숩)은 전혀 다른 라이브러리! 사용하지 않는다.
# as 알리아스 별칭
from bs4 import BeautifulSoup as bs
```

```python
# bs아 내가 가지고 있는 글자 정보를 html 정보로 바꿔줘~
# bs("내가 가지고 있는 정보", "파싱 방법(어떻게 처리하는 지)")
soup = bs(res.text, "lxml") # BeautifulSoup 객체화 완료! : html 소스로 바뀌었음
```

```python
print(soup)
```

```python
# 네이버 페이지에서 카페, 블로그, 지식인 글자를 수집해!
# 데이터에서 a태그만 수집해줘~
# html 소스.select("선택자")
data = soup.select("a.nav")

# But, 계층 선택자, 동적 태그 등 예외가 너무 많음
```

```python
print(data)
```

```python
for i in data :
    print(i.text)
```

```python
for i in data[:3] :
    print(i.text)
```

# 2. 뉴스 타이틀 수집

```python
res1 = req.get('https://search.naver.com/search.naver?where=news&sm=tab_jum&query=%EC%9E%A5%EB%A7%88')
soup1 = bs(res1.text, 'lxml')
data1 = soup1.select("a.news_tit")
for i in data1 :
    print(i.text)
```

- 모바일 페이지는 클래스명이 다름!!
- 이런 예외가 발생할 수 있으니 무조건 PC용 웹 페이지를 사용하자!


# 3. 네이버 월요 웹툰 제목 크롤링

```python
# 네이버 월요 웹툰 제목 크롤링
res2 = req.get('https://comic.naver.com/webtoon/weekdayList?week=mon')
soup2 = bs(res2.text, 'lxml')
data2 = soup2.select("dt")
for i in data2 :
    print(i.text.strip())
```
